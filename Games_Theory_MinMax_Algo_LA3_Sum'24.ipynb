```python
#Task1
import random

def MinMaxStrategy(level, index, is_maximizing, score_list, lower_bound_A, upper_bound_B):
    if level == 5:
        return score_list[index]

    if is_maximizing:
        current_max = float('-inf')

        for x in range(2):
            current_score = MinMaxStrategy(level + 1, index * 2 + x, False, score_list, lower_bound_A, upper_bound_B)
            current_max = max(current_max, current_score)
            lower_bound_A = max(lower_bound_A, current_score)

            if upper_bound_B <= lower_bound_A:
                break

        return current_max
    else:
        current_min = float('inf')

        R_count = 3

        for x in range(R_count-1):
            current_score = MinMaxStrategy(level + 1, index * 2 + x, True, score_list, lower_bound_A, upper_bound_B)
            current_min = min(current_min, current_score)
            upper_bound_B = min(upper_bound_B, current_score)

            if upper_bound_B <= lower_bound_A:
                break

        return current_min

def start_battle(first_player):
    score_list = []

    for i in range(2**5):

       score = random.choice([-1, 1])

       score_list.append(score)

    if first_player == 0:
        result = MinMaxStrategy(0, 0, True, score_list, float('-inf'), float('inf'))
    else:
        result = MinMaxStrategy(0, 0, False, score_list, float('-inf'), float('inf'))

    return 'Scorpion' if result == -1 else 'Sub-Zero'

def KombatRules(initial_turn):
    total_rounds = 3
    round_outcomes = []

    current_player = initial_turn

    for _ in range(total_rounds):
        winner = start_battle(current_player)
        round_outcomes.append(winner)

        current_player = 1 - current_player

    scorpion_wins = round_outcomes.count('Scorpion')
    subzero_wins = round_outcomes.count('Sub-Zero')

    if scorpion_wins > subzero_wins:
        game_winner = 'Scorpion'
    else:
        game_winner = 'Sub-Zero'

    return {
        'Game Winner': game_winner,
        'Rounds Played': total_rounds,
        'Round 1 Winner': round_outcomes[0],
        'Round 2 Winner': round_outcomes[1],
        'Round 3 Winner': round_outcomes[2]
    }

first_turn_input = input("Enter starting player: ")

try:
    P_1 = int(first_turn_input)
    if P_1 not in [0, 1]:
        raise ValueError("Not VALID")
except ValueError:
    print("Invalid input! Defaulting to Scorpion (0).")
    P_1 = 0

final_result = KombatRules(P_1)

for key, value in final_result.items():
    print(f"{key}: {value}")
```

```python
#Task2
def MinimaxTraversal(node, depth, Val_of_A, Val_OF_B, maximize):
    if depth == 0 or not node.children:
        return node.val

    if maximize:
        max_check = float('-inf')
        for child in node.children:
            evaluation = MinimaxTraversal(child, depth - 1, Val_of_A, Val_OF_B, False)
            max_check = max(max_check, evaluation)
            Val_of_A = max(Val_of_A, evaluation)
            if Val_OF_B<= Val_of_A:
                break
        return max_check
    else:
        min_check = float('inf')
        for child in node.children:
            evaluation = MinimaxTraversal(child, depth - 1, Val_of_A, Val_OF_B, True)
            min_check = min(min_check, evaluation)
            Val_OF_B= min(Val_OF_B, evaluation)
            if Val_OF_B<= Val_of_A:
                break
        return min_check

class PacNode:
    def __init__(self, val, children=[]):
        self.val = val
        self.children = children

def PacmanAdventure(difficulty):
    leaf_vaL = [3, 6, 2, 3, 7, 1, 2, 0]
    leaf_nodes = [PacNode(val) for val in leaf_vaL]

    level_2 = []

    for i in range(0, len(leaf_nodes), 2):
        new_node = PacNode(None, [leaf_nodes[i], leaf_nodes[i + 1]])
        level_2.append(new_node)

    level_1 = []

    for i in range(0, len(level_2), 2):
        new_node = PacNode(None, [level_2[i], level_2[i + 1]])
        level_1.append(new_node)

    root = PacNode(None, level_1)

    minimax_value = MinimaxTraversal(root, 3, float('-inf'), float('inf'), True)

    left_max = max(leaf_vaL[:4]) - difficulty
    right_max = max(leaf_vaL[4:]) - difficulty

    if left_max > minimax_value or right_max > minimax_value:
        if left_max > right_max:
            print(f"The new minimax value is {left_max}. Pacman goes left and uses dark magic")
        else:
            print(f"The new minimax value is {right_max}. Pacman goes right and uses dark magic")
    else:
        print(f"The minimax value is {minimax_value}. Pacman does not use dark magic")

user_input = int(input("Enter a number: "))
PacmanAdventure(user_input)
```
