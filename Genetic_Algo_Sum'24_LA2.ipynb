```python
#Task 1
import random

def calculate_overlap(schedule, Num_of_Courses, TimeSlots):
    o_panelty = 0
    for i in range(TimeSlots):
        indivisual_slot = schedule[i * Num_of_Courses:(i + 1) * Num_of_Courses]
        course_one_slot = 0
        for i in indivisual_slot:
            if i=="1":
                course_one_slot+=1
        if course_one_slot > 1:
            o_panelty += course_one_slot - 1
    return o_panelty

def calculate_consistency(schedule, num_of_courses, TimeSlots):
    c_panelty = 0
    for i in range(num_of_courses):
        course_scheduled = 0

        for j in range(TimeSlots):
            index = j * num_of_courses + i
            course_scheduled += int(schedule[index])


        if course_scheduled != 1:
            c_panelty += abs(course_scheduled - 1)

    return c_panelty

def fitness(schedule, num_of_courses, TimeSlots):
    return - (calculate_overlap(schedule, num_of_courses, TimeSlots)+ calculate_consistency(schedule, num_of_courses, TimeSlots))

def random_parents(population, num_of_courses, TimeSlots):
    parent1 = random.choice(population)
    parent2 = random.choice(population)
    return parent1, parent2

def crossover(parent1, parent2, num_of_courses, TimeSlots):
    crossover_point = random.randint(1, TimeSlots*num_of_courses - 1)

    return parent1[:crossover_point] + parent2[crossover_point:], parent2[:crossover_point] + parent1[crossover_point:]
def mutate(schedule, num_of_courses, TimeSlots):
    mutation_ind = random.randint(0, num_of_courses * TimeSlots - 1)
    schedule_as_list = list(schedule)
    if schedule_as_list[mutation_ind]=="1":
        schedule_as_list[mutation_ind]="0"
    else:
        schedule_as_list[mutation_ind]="1"
    mutated=''.join(schedule_as_list)
    return mutated
def population_create(num_of_courses,TimeSlots,population_size):
    population=[]
    for _ in range(population_size):
        individual = []
        for _ in range(num_of_courses * TimeSlots):
            individual.append(random.choice('01'))

        population.append(''.join(individual))
    return population


def genetic_algorithm_for_BracU_Schedule(N, T, population_size, max_iterations):
    population = population_create(N,T,population_size)
    for _ in range(max_iterations):
        new_population = []
        for _ in range(population_size // 2):
            parent1, parent2 = random_parents(population, N, T)
            cross_child1, cross_child2 = crossover(parent1, parent2, N, T)
            mutated_child1 = mutate(cross_child1, N, T)
            mutated_child2 = mutate(cross_child2, N, T)
            new_population.extend([mutated_child1, mutated_child2])
        population = new_population
        fitness_list = []
        for schedule in population:
            fitness_list.append(fitness(schedule, N, T))
        max_fitness = max(fitness_list)
        best_schedule_index = fitness_list.index(max_fitness)
        best_schedule = population[best_schedule_index]

        if max(fitness_list) == 0:
            break
    return best_schedule, max(fitness_list)

N, T = map(int, input("Enter N and T : ").split())
courses = [input("Course name: ") for _ in range(N)]
if N>T:
    print("Error : N should be less than T")
else:
    best_schedule, best_fitness = genetic_algorithm_for_BracU_Schedule(N, T, 100, 1000)
    print(best_schedule)
    print(best_fitness)
```
